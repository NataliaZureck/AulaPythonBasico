## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t

print("p:", p)
print("q:", q)
print("r:", r)
print("s:", s)
print("t:", t)

if resultado:
    print("Resultado: True")
else:
    print("Resultado: False")
```

Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

Em seguida, foi criado um conjunto de números aleatórios com distribuição normal utilizando a função np.random.normal(), definindo a média 10 com loc, o desvio padrão 3 com scale e a quantidade de 5 valores aleatorios com o size. Os valores foram armazenados em um array e identificados quais os valores eram maiores que 10. esses valores foram contados e impressos no resultado.

## 04 – Escreva o programa em Python:

```python
import numpy as np

array = np.random.normal(loc=10, scale=3, size=5)

valores_maiores = array[array > 10]
quantidade = (array > 10).sum()

print("Array:", array)
print("Valores > 10:", valores_maiores)
print("Quantidade:", quantidade)
```

## No exercício 7 da aula 6, foi instanciado um objeto da classe DataFrame. Exporte (usando nome_data_frame.to_csv(‘nome_arquivo.csv’)) para .csv estes dados e os importe (com pd.read_csv(‘nome_arquivo.csv’) aqui. Edite o objeto para que os index sejam os nomes dos genes e para que as colunas sejam facilmente identificadas. Usando o número de caracteres do primeiro gene, quero saber a informação gênica. Caso o nome tenha entre 2 e 4 caracteres inclusive, posso saber as informações do gene? Porém, caso tenha menos de 2 caracteres, posso saber todos os valores nas instâncias onde os números de caracteres dos nomes são superiores a 4? Ainda, caso tenha mais de 4 caracteres, apenas me informe quem é superior.


## 07 – Descreva o racional da sua resposta:

R.: A DataFrame foi exportada para um arquivo CSV utilizando o método to_csv() no notebook original. Em seguida os dados foram importados com o método read_csv(), utilizando o parâmetro index_col=0 para manter os índices. E então foi identificado o primeiro gene do DataFrame e calculado o número de caracteres do seu nome utilizando a função len(). Com isso, foi criada uma estrutura condicional para analisar se o nome do gene possuia entre 2 e 4 caracteres, nesse caso ele devolve as informações desse gene. Se fosse menos de 2 caracteres devolvia dos genes que tivessem mais de 4 e se fosse maior que 4 devolvia  os genes com mais de 4 caracteres.

## 08 – Desenhe o fluxograma da resposta:

<img src="./edição dataframe_10_9.png" width="400" />

## 09 – Escreva o programa em Python:

```python
import pandas as pd
df = pd.read_csv('genes.csv', index_col=0)
gene_1 = df.index[0]
num_caracteres = len(gene_1)
if 2 <= num_caracteres <= 4:
    print("Gene com nome entre 2 e 4 caracteres:")
    print(df.iloc[0])
elif num_caracteres < 2:
    print("Genes com nome menor que 2 caracteres:")
    print(df[df.index.str.len() < 2])
else:
    print("Genes com nome maior que 4 caracteres:")
    print(df[df.index.str.len() > 4])
```
