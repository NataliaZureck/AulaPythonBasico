## 02 – Obtenha os resultados e retorne o código do objeto em Python dos seguintes cálculos:

b) lista_a * 2

c) lista_a + lista_b

d) lista_a / (lista_b * lista_c)

```python
import numpy as np
lista_a = [2.0, 3.0, 5.0] + [ 10, 20, 30]
lista_a = np.array(lista_a)
print(lista_a)
lista_b = lista_a * 2
print(lista_b)
lista_c = lista_a + lista_b
print(lista_c)
lista_d = lista_a / (lista_b * lista_c)
print(lista_d)
```

## 06 – Insira as médias das colunas no final do array do exercício 4.

**Contexto:**
- No exercício 4, você criou um array numpy com 5 linhas e 4 colunas (totalizando 20 valores)
- No exercício 5, você calculou a média de cada uma das 4 colunas (resultando em uma lista com 4 valores)

**Tarefa:**
1. Use o atributo `shape` do array do exercício 4 para calcular o número total de elementos.
2. Insira a lista de médias (do exercício 5) no final de outro array criado
3. Recupere e exiba o array resultante

## 07 – Remodele (reshape) o array do exercício anterior, para que contenha 4 colunas. Converta este objeto da classe array para DataFrame. Arredonde todos os valores para até 3 casas decimais. Recupere a DataFrame.

```python
import pandas as pd
array_9 = array_8.reshape(6, 4)
dataframe1 = pd.DataFrame(array_9).round(3)
print(dataframe1)
```

## 08 – A última instancia desta DataFrame deve conter as médias dos valores contidos por coluna, o que foi determinado nos exercícios anteriores. Instancie um objeto contendo as médias dos valores em cada linha. Insira este objeto como uma nova coluna na DataFrame. Recupere a DataFrame.

```python
medias_dataframe = dataframe1.mean(axis=1)

dataframe1['media_linhas'] = medias_dataframe

print(dataframe1)
```

## 09 – Usando o método dos objetos de classe DataFrame, describe(), compare os resultados contidos na última linha e última coluna da DataFrame.

```python
ultima_linha = dataframe1.iloc[-1]
ultima_coluna = dataframe1['media_linhas']

print("Última linha:")
print(ultima_linha.describe())

print("\nÚltima coluna:")
print(ultima_coluna.describe())
```
