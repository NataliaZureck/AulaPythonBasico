## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import pandas as pd
df = pd.read_csv('data.csv')
num_linhas = len(df)
classes_colunas = list(df.dtypes)
quantidade_elementos = list(df.count())
colunas = [0, 3, 9]
medias = [df.iloc[:, i].mean() for i in colunas]
desvios = [df.iloc[:, i].std() for i in colunas]
terceiro_quartil = [df.iloc[:, i].quantile(0.75) for i in colunas]
print(medias, desvios, terceiro_quartil)
resultados = {
    'num_linhas': num_linhas,
    'classes_colunas': classes_colunas,
    'quantidade_elementos': quantidade_elementos,
    'medias': medias,
    'desvios': desvios,
    'terceiro_quartil': terceiro_quartil
}
print(resultados)
```
