Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: primeiro, foi criada uma função para obter um array de valores aleatórios com distribuição normal, definindo os padrões de média 10 e desvio padrão 3. Em seguida, foi criada uma segunda função para analisar os valores desse array, filtrando aqueles que são superiores a 10 e verificando a quantidade.

## 04 – Escreva o programa em Python:

```python
import numpy as np
def gerar_array():
    return np.random.normal(loc=10, scale=3, size=5)

def analisar_valores(array):
    valores_maiores = array[array > 10]
    quantidade = (array > 10).sum()
    return valores_maiores, quantidade

array = gerar_array()
valores, qtd = analisar_valores(array)
print("Array:", array)
print("Valores > 10:", valores)
print("Quantidade:", qtd)
```

```python
def calculo_media(valores):
    media = sum(valores) / len(valores)
    return media
lista = [10, 20, 30, 40, 50]

resultado = calculo_media(lista)
print("Média:", resultado)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python

def dividir(a, b):
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        return "Os dois valores devem ser numéricos"
    
    if b == 0:
        return "Não é permitida a divisão por 0"
    resultado = a / b
    
    return f"Tipos: {type(a).__name__}, {type(b).__name__} | Resultado: {resultado}"
print(dividir(10, 2))
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
import numpy as np
def dividir_compostos(a, b):
    if not isinstance(a, (list, np.ndarray)) or not isinstance(b, (list, np.ndarray)):
        return "os dois precisam ser listas ou arrays."
    a = np.array(a)
    b = np.array(b)
    if a.shape != b.shape:
        return "Os objetos devem ter o mesmo tamanho."
    if (b == 0).any():
        return "Divisão por zero não é permitida."
    resultado = a / b
    return f"Tipos: {type(a).__name__}, {type(b).__name__} | Resultado: {resultado}"
print(dividir_compostos([10, 20, 30], [2, 4, 5]))
```
